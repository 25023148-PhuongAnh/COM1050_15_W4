# A3

```python
def nguyento(a):
    if a<=1:
        return "No"
    for i in range(2,int(a**0.5)+1):
        if a%i==0:
            return "No"
    return "Yes"
n=int(input())
nguyento(n)
```

# A4

```python
n=int(input())
def sohoanhao(a):
    tong=0
    for i in range(1,a):
        if a%i==0:
            tong=tong+i
    if tong==a:
        return "YES"
    else:
        return "NO"
sohoanhao(n)
```

# A5

```python
def ktvt(n,k):
    for i in range(len(n)):
        if n[i]==k:
            return i
    return -1
a,k=map(int,input().split())
n=[]
for i in range(0,n):
    x=int(input())
    n.append(x)
ktvt(n,k)
```

# A7

```python
def cal(a,b,c):
    if c=="+":
        return a+b
    elif c=="-":
        return a-b
    elif c=="*":
        return a*b
    elif c=="/":
        return a/b
n1,n2,oper=input().split()
n1=float(n1)
n2=float(n2)
ketqua=cal(n1,n2,oper)
round(ketqua,2)
```

```python
n=int(input())
def ttong(a):
    tong = 0
    while a>0:
        tong+=a%10
        a//=10
    return tong
ttong(n)
```
