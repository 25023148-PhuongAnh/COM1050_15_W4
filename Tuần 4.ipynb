## W4A9

```python
n=int(input())
ketqua=[]
k=1
while k*k <= n:
    sq=k*k
    s = str(sq)
    if len(set(s)) == len(s):
        ketqua.append(s)
    k+=1
if ketqua:
    print(" ".join(ketqua))
else:
    print("no number")
```
