## A1

```python
n = int(input("Nhập n: "))
tong = 0
for i in range(1,n+1):
    tong += i
print("Tổng là:", tong)
```

## A2

```python
while True:
    n = int(input("Nhập số nguyên dương: "))
    if n > 0:
        break
if n<2:
    print("Không phải số nguyên tố")
else:
    for i in range(2, n):
        if n % i == 0:
            print("Không phải số nguyên tố")
            break
    else: 
            print("Số nguyên tố")
```
